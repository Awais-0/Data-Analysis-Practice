```python
import requests
from bs4 import BeautifulSoup
import csv

response = requests.get('https://www.passiton.com/inspirational-quotes')

soup = BeautifulSoup(response.text)
```

```python
quotes = []

for div in soup.find_all('div', class_ = 'text-center mb-8'):
    quotes.append(div.find('p', class_ = 'text-white').text)

print(quotes)
```
