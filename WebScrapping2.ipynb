```python
import requests
from bs4 import BeautifulSoup

URL = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
headers = {
    "User-Agent": "AwaisDataFetcher/1.0 (mailto:awais@example.com)",
    "Accept": "text/html"
}
response = requests.get(URL, headers=headers)
soup = BeautifulSoup(response.text, 'html')

# print(soup.prettify())
```

```python
my_table = soup.find_all('table')[0]
my_headers = [title.text.strip() for title in my_table.find_all('th')]

my_datarows = [[td.text.strip() for td in datarow.find_all('td')] for datarow in my_table.find_all('tr')[1:]]
print(my_datarows)
```

```python
import pandas as pd
```

```python


data_frame = pd.DataFrame(columns=my_headers, data=my_datarows)
data_frame
```
